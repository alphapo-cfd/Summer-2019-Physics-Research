```python
u = u_ana(x,0)
q = d_dx_periodic(u,a,dx)
plt.plot(x, u, label = 'u')
plt.plot(x, q, label = 'q', linestyle ='--')
plt.legend()
```

```python
t = 0 
t_target = 1.
nt = int((t_target-t)/dt)
print (nt)
```
