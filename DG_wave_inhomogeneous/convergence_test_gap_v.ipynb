```python
import numpy as np
from matplotlib import pyplot,cm as plt,cm
from scipy.interpolate import *
from ReferenceElement import *
import os
import imageio
from DG_functions_wave_inhomogeneous import *
```

```python
N_range = np.array([2,4,6,8,10])
K_range = np.array([2,4,6,8,10])
CF = 2.**(-7)
start = 0.
end = 1.
a1 = 1.
a2 = -1.
alpha = 0.
def source_A(t):
    #return 5+0*t
    return np.sin(5*np.pi*t)
def source_B(t):
    return np.sin(5*np.pi*t)
def source_B_dot(t):
    return 5*np.pi*np.cos(5*np.pi*t)
def source_for_v(t):
    return source_A(t) + source_B_dot(t)
```

```python
error_same_K = []
for N in N_range:
    error_same_N = []
    for K in K_range:
        h = (end-start)/K
        reference_element = ReferenceElement(N)
        reference_interval = reference_element[0]
        M_inv = reference_element[1]*2/h
        M_inv_S = reference_element[2]*2/h
        x = get_x_elements(start,end, K, reference_interval)
        dx_min = get_dx_min(x)
        u = np.zeros_like(x)
        v = np.zeros_like(x)
        t = 0.
        t_target = .4
        dt = CF*dx_min
        nt = int((t_target-t)/dt)
        for n in range(nt):
            un = u.copy()
            vn = v.copy()
            u = RK4_Step(dt,DG_du_dt_radiative,u, K,N,t,a1,alpha,M_inv, M_inv_S,source_B,v)
            v = RK4_Step(dt,DG_du_dt_radiative,v, K,N,t,a2,alpha,M_inv, M_inv_S,source_for_v)
            t += dt 
        gap_v = v[int(K/2-1)][-1]-v[int(K/2)][0]
        gap_v_analytical = source_for_v(t)
        error = np.log10(np.abs((gap_v-gap_v_analytical)/gap_v_analytical))
        error_same_N.append(error)
    error_same_K.append(error_same_N)
    plt.plot(K_range,error_same_N, label = 'N = '+str(N))
    plt.scatter(K_range,error_same_N)
```

```python
for i in range(len(error_same_K)):
    plt.plot(K_range, error_same_K[i], label = 'N = '+str(N_range[i]), marker = 'o')
plt.ylabel('log10 error')
plt.xlabel('K')
plt.legend()
```
